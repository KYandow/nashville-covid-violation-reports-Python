```python
hubCovid_df = hub_df.loc[hub_df['Subrequest Type'] =='COVID-19 Violations']
hubCovid_df
```

```python
# explore frequency per day, location over time, length of time from open to close
```

```python
hubCovid_df = hubCovid_df.rename(columns = {'Date / Time Opened': 'date_opened', 'Date / Time Closed' : 'date_closed'})
hubCovid_df
```

```python
hubCovid_df['date_opened'] = pd.to_datetime(hubCovid_df['date_opened'], yearfirst = True)
hubCovid_df['date_closed'] = pd.to_datetime(hubCovid_df['date_closed'], yearfirst = True)
```

```python
hubCovid_df['Month'] = hubCovid_df['date_opened'].dt.month
hubCovid_df.head()
```

```python
hubCovid_df['Day'] = hubCovid_df['date_opened'].dt.day
hubCovid_df
```

```python
hubCovid_df['Resolution Time'] = hubCovid_df['date_closed']-hubCovid_df['date_opened']
hubCovid_df.head()
```

```python
max_resolution_time = max((hubCovid_df['date_closed'] - hubCovid_df['date_opened']).dt.days)
min_resolution_time = min((hubCovid_df['date_closed'] - hubCovid_df['date_opened']).dt.days)

print(f"The longest time to resolution was {max_resolution_time} days, while the shortest was {min_resolution_time} days")
```

```python
mean_resolution_time = np.mean(hubCovid_df['Resolution Time'])
print(mean_resolution_time)
```
